```python

# Extract epochs
epochs = list(data.keys())

# Extract AP and AR metrics
overall_ap_metrics = ['AP_50_95', 'AP_50', 'AP_75']
size_ap_metrics = ['AP_50_95_medium', 'AP_50_95_large']
overall_ar_metrics = ['AR_1', 'AR_10', 'AR_100']
size_ar_metrics = ['AR_100_medium', 'AR_100_large']

def plot_metrics(metrics, title, ylabel, vertical_line_epochs=None, annotation_texts=None):
    plt.figure(figsize=(14, 7))
    
    # Initialize a variable to store the global maximum value for annotation positioning
    global_max_value = float('-inf')
    
    for metric in metrics:
        values = [data[epoch][metric] for epoch in epochs]
        plt.plot(epochs, values, marker='o', label=metric)
        # Update the global maximum value if the current metric's max is higher
        max_value = max(values)
        if max_value > global_max_value:
            global_max_value = max_value
    
    # Add vertical lines and annotations
    if vertical_line_epochs is not None and annotation_texts is not None:
        for i in range(len(vertical_line_epochs)):
            plt.axvline(x=vertical_line_epochs[i], color='red', linestyle='--', linewidth=2)
            # Position the text just above the grid of the plot
            plt.text(vertical_line_epochs[i], global_max_value + (global_max_value * 0.05),  # Slightly above the maximum value
                     annotation_texts[i], color='red', ha='center')
    
    plt.title(title, pad=20)  # Adjust the pad value to move the title higher
    plt.xlabel('Epochs')
    plt.ylabel(ylabel)
    plt.legend()
    plt.grid(True)
    plt.show()

# Example usage with multiple vertical lines and annotation texts
plot_metrics(overall_ap_metrics, 'Average Precision (AP) Metrics - Overall', 'AP', vertical_line_epochs=[5, 6, 13, 25, 26], annotation_texts=["↓ lr", "↓ lr", 'New training', "↓ lr", "↓ lr"])

# Plot Average Precision (Small, Medium, Large) with a single vertical line and annotation
plot_metrics(size_ap_metrics, 'Average Precision (AP) Metrics - Medium, Large', 'AP', vertical_line_epochs=[5, 6, 13, 25, 26], annotation_texts=["↓ lr", "↓ lr", 'New training', "↓ lr", "↓ lr"])

# Plot Average Recall (Overall) with a single vertical line and annotation
plot_metrics(overall_ar_metrics, 'Average Recall (AR) Metrics - Overall', 'AR', vertical_line_epochs=[5, 6, 13, 25, 26], annotation_texts=["↓ lr", "↓ lr", 'New training', "↓ lr", "↓ lr"])

# Plot Average Recall (Small, Medium, Large) with a single vertical line and annotation
plot_metrics(size_ar_metrics, 'Average Recall (AR) Metrics - Medium, Large', 'AR', vertical_line_epochs=[5, 6, 13, 25,26], annotation_texts=["↓ lr", "↓ lr", 'New training', "↓ lr", "↓ lr"])
```
