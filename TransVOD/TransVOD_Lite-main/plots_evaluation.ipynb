```python
# Extract epochs
epochs = list(data.keys())

# Extract AP and AR metrics
overall_ap_metrics = ['AP_50_95', 'AP_50', 'AP_75']
size_ap_metrics = ['AP_50_95_medium', 'AP_50_95_large']
overall_ar_metrics = ['AR_1', 'AR_10', 'AR_100']
size_ar_metrics = ['AR_100_medium', 'AR_100_large']

# Function to plot metrics
def plot_metrics(metrics, title, ylabel):
    plt.figure(figsize=(14, 7))
    for metric in metrics:
        values = [data[epoch][metric] for epoch in epochs]
        plt.plot(epochs, values, marker='o', label=metric)
    plt.title(title)
    plt.xlabel('Epochs')
    plt.ylabel(ylabel)
    plt.legend()
    plt.grid(True)
    plt.show()

# Plot Average Precision (Overall)
plot_metrics(overall_ap_metrics, 'Average Precision (AP) Metrics - Overall', 'AP')

# Plot Average Precision (Small, Medium, Large)
plot_metrics(size_ap_metrics, 'Average Precision (AP) Metrics - Small, Medium, Large', 'AP')

# Plot Average Recall (Overall)
plot_metrics(overall_ar_metrics, 'Average Recall (AR) Metrics - Overall', 'AR')

# Plot Average Recall (Small, Medium, Large)
plot_metrics(size_ar_metrics, 'Average Recall (AR) Metrics - Small, Medium, Large', 'AR')
```
